```python
import pandas as pd
data={
    "Department":["IT","IT","HR","HR"],
    "Name":["A","B","c","D"],
    "Marks":[80,85,75,90]
}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Grouping
print("\n average marks by department:")
print(df.groupby("Department")["Marks"].sum())
```
