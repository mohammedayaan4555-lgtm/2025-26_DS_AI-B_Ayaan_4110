```python
#Plotting
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("X-Y graph")
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
subjects=["science","physics","cs","maths"]
marks=[80,75,85,90]
plt.bar(subjects,marks)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.title("BAR CHART")
plt.show()
```

```python
#Scattered graph
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,20,25,30]

plt.scatter(x,y)
plt.title("Scattered Graph")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
data=[10,12,15,18,20,25,30,35,40]

plt.hist(data,bins=5,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

labels = ["Maths", "Physics", "Chemistry", "CS"]
marks = [80, 75, 85, 90]

plt.pie(marks, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
