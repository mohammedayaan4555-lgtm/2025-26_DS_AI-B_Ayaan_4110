```python
# Insertion Sort

n = int(input("enter num of elements: "))

arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}: ")))

print("\nOriginal array:", arr)

# Insertion sort logic
for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1

    arr[j+1] = key

print("Sorted array:", arr)
```
