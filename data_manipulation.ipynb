```python
#Data Manipulation
import pandas as pd

#Create DataFrame
data={
    "Name":["A","B","C"],
    "Age":[21,22,23],
    "Marks":[70,80,90]
}

df=pd.DataFrame(data)

print("original Data:")
print(df)


#Add new column
df["Bonus"]=5
print("\nAfter adding Bonus:")
print(df)

#Increasing marks
df["Marks"]=df["Marks"]+2
print("\nAfter increasing marks:")
print(df)

#Change value
df.loc[0,"Age"]=21
print("\nAfter changing value:")
print(df)

#Rename column
df.rename(columns={"Total":"final"},inplace=True)
print("\n After renaming total to final:")
print(df)

#Remove column
df.drop("Bonus",axis=1,inplace=True)
print("\n After removing bonus:")
print(df)
```
