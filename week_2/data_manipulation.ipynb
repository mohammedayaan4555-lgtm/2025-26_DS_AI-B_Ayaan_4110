```python
# data manipulaton
import pandas as pd

data={                       #creat data frame
    "Name":["A","B","C"],
    "Age":[21,22,23],
    "Marks":[70,80,90]
}
df=pd.DataFrame(data)
print("original Data:")
print(df)
df["Bonus"]=5
print("\nAfter adding Bonus:")
print(df)
df["Marks"]=df["Marks"]+2
print("\nAfter increasing marks:")
print(df)
df.loc[0,"Age"]=21
print("\nAfter changing value:")
print(df)
df.rename(columns={"Total":"final"},inplace=True)
print("\n After renaming total to final:")
print(df)
df.drop("Bonus",axis=1,inplace=True)
print("\n After removing bonus:")
print(df)
```
