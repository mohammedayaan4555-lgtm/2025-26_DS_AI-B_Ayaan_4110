Ranking

```python
import pandas as pd
data={
    "name":["A","B","C"],
    "age":[21,22,23],
    "marks":[85,90,80]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
df["ranks"]=df["marks"].rank(ascending=False)
print("\nafter ranking:")
print(df)
```
