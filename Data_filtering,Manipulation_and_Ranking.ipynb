```python
#filtering
import pandas as pd

#creating data
data={
    "ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eva"],
    "Age":[22,25,23,24,26],
    "Marks":[85,78,92,88,90]
}

#Creating DataFrame
df=pd.DataFrame(data)
print("Original DataFrame")
print(df)

#Students with marks > 85
print("\nStudents with marks > 85:")
print(df[df["Marks"]>85])

#Students with age < 25
print("\nStudents with age < 25:")
print(df[df["Age"] < 25])

#Students with age > 23 AND marks >= 88
print("\nStudents with age > 23 AND marks >= 88:")
print(df[(df["Age"] > 23) & (df["Marks"] >= 88)])

#Students with marks > 90 OR age < 23
print("\nStuents with marks > 90 OR age < 23:")
print(df[(df["Marks"] > 90) | (df["Age"] < 23)])
```

```python
#Data Manipulation
import pandas as pd

#Create DataFrame
data={
    "Name":["A","B","C"],
    "Age":[21,22,23],
    "Marks":[70,80,90]
}

df=pd.DataFrame(data)

print("original Data:")
print(df)


#Add new column
df["Bonus"]=5
print("\nAfter adding Bonus:")
print(df)

#Increasing marks
df["Marks"]=df["Marks"]+2
print("\nAfter increasing marks:")
print(df)

#Change value
df.loc[0,"Age"]=21
print("\nAfter changing value:")
print(df)

#Rename column
df.rename(columns={"Total":"final"},inplace=True)
print("\n After renaming total to final:")
print(df)

#Remove column
df.drop("Bonus",axis=1,inplace=True)
print("\n After removing bonus:")
print(df)
```

```python
#ranking
import pandas as pd
data={
    "name":["A","B","C"],
    "age":[21,22,23],
    "marks":[85,90,80]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
df["ranks"]=df["marks"].rank(ascending=False)
print("\n after ranking:")
print(df)
```
