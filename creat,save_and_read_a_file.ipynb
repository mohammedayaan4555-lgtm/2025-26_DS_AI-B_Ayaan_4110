4.Write the code to create, save, and read a file from the local drive and
perform text operations such as counting the number of lines, words, and
characters.

```python

# Create and write to a file
sample_file = "sample_text.txt"

with open(sample_file, "w") as file:
    file.write("Hello World\n")
    file.write("This is a Python file handling example\n")
    file.write("Counting lines words and characters")

print("File created and data written!")

# Read file and perform operations
with open(sample_file, "r") as file:
    text = file.read()

# Count lines
lines = text.split("\n")
line_count = len(lines)

# Count words
word_count = len(text.split())

# Count characters
char_count = len(text)

print("Number of lines:", line_count)
print("Number of words:", word_count)
print("Number of characters:", char_count)
```
