```python


# 1. Creating lists
nums = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.5, True]
empty = []

print("Original list:", nums)

# 2. Accessing elements
print("First element:", nums[0])
print("Last element:", nums[-1])

# 3. Slicing
print("Slice [1:4]:", nums[1:4])
print("Slice [:3]:", nums[:3])
print("Slice [::2]:", nums[::2])

# 4. Modifying elements
nums[0] = 10
print("After modification:", nums)

# 5. Adding elements
nums.append(6)
print("After append:", nums)

nums.insert(1, 20)
print("After insert:", nums)

nums.extend([7, 8, 9])
print("After extend:", nums)

# 6. Removing elements
nums.remove(20)
print("After remove:", nums)

popped = nums.pop()
print("After pop:", nums, "| Popped:", popped)

nums.pop(0)
print("After pop at index 0:", nums)

# 7. Deleting elements
del nums[1]
print("After del:", nums)

# 8. Clearing list
temp = nums.copy()
temp.clear()
print("After clear:", temp)

# 9. Searching
print("Index of 4:", nums.index(4))
print("Count of 3:", nums.count(3))

# 10. Sorting
nums.sort()
print("Sorted:", nums)

nums.sort(reverse=True)
print("Reverse sorted:", nums)

# 11. Reversing
nums.reverse()
print("Reversed:", nums)

# 12. Copying
copy1 = nums.copy()
copy2 = list(nums)
print("Copies:", copy1, copy2)

# 13. Length
print("Length:", len(nums))

# 15. Iteration
print("Looping through list:")
for item in nums:
    print(item)

# 16. List comprehension
squares = [x * x for x in nums]
print("Squares:", squares)

# 17. Nested lists
matrix = [[1, 2], [3, 4], [5, 6]]
print("Nested list:", matrix)
print("Access nested element:", matrix[1][0])

# 18. Min, Max, Sum
print("Min:", min(nums))
print("Max:", max(nums))
print("Sum:", sum(nums))

# 19. Joining list elements
words = ["Python", "is", "fun"]
sentence = " ".join(words)
print("Joined string:", sentence)


```

```python


# 1. Creating tuples
empty = ()
single = (5,)  # note the comma
numbers = (1, 2, 3, 4, 5)
mixed = (1, "hello", 3.5, True)

print("Original tuple:", numbers)
print("Empty tuple:", empty)
print("Single element tuple:", single)

# 2. Accessing elements
print("First element:", numbers[0])
print("Last element:", numbers[-1])

# 3. Slicing
print("Slice [1:4]:", numbers[1:4])
print("Slice [:3]:", numbers[:3])
print("Slice [::2]:", numbers[::2])

# 4. Tuple concatenation
t1 = (1, 2)
t2 = (3, 4)
t3 = t1 + t2
print("Concatenated tuple:", t3)

# 5. Tuple repetition
t4 = t1 * 3
print("Repeated tuple:", t4)



# 7. Iteration
print("Looping through tuple:")
for item in numbers:
    print(item)





# 10. Built-in functions
print("Length:", len(numbers))
print("Max:", max(numbers))
print("Min:", min(numbers))
print("Sum:", sum(numbers))

# 11. Tuple methods
# Tuples have only two methods: count() and index()
print("Count of 2:", numbers.count(2))
print("Index of 4:", numbers.index(4))

# 12. Nested tuples
nested = ((1, 2), (3, 4), (5, 6))
print("Nested tuple:", nested)
print("Access nested element:", nested[1][0])

# 13. Converting between list and tuple
lst = list(numbers)
print("Tuple to list:", lst)
t_from_list = tuple(lst)
print("List back to tuple:", t_from_list)

# 14. Immutability demonstration
try:
    numbers[0] = 10
except TypeError as e:
    print("Cannot modify tuple:", e)
```

```python


# 1. Creating sets
empty_set = set()           # empty set
numbers = {1, 2, 3, 4, 5}  # set with elements
duplicates = {1, 2, 2, 3}  # duplicates are removed automatically
print("Empty set:", empty_set)
print("Numbers set:", numbers)
print("Duplicates removed:", duplicates)

# 2. Adding elements
numbers.add(6)
print("After add(6):", numbers)

numbers.update([7, 8, 9])  # add multiple elements
print("After update([7,8,9]):", numbers)

# 3. Removing elements
numbers.remove(9)
print("After remove(9):", numbers)

numbers.discard(10)
print("After discard(10):", numbers)

popped = numbers.pop()
print("After pop():", numbers, "| Popped:", popped)

# 4. Clearing set
temp = numbers.copy()
temp.clear()
print("After clear():", temp)

# 5. Set operations
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Set A:", A)
print("Set B:", B)

print("Union (A | B):", A | B)
print("Intersection (A & B):", A & B)
print("Difference (A - B):", A - B)
print("Symmetric Difference (A ^ B):", A ^ B)
# 8. Iteration
print("Looping through set A:")
for item in A:
    print(item)



```

```python


# 1. Creating a dictionary
fruits = {"apple": 10, "banana": 5}
print("Original dictionary:", fruits)

# 2. Accessing elements
print("Number of apples:", fruits["apple"])
print("Number of bananas (using get):", fruits.get("banana"))

# 3. Adding/updating elements (easy example)
fruits["orange"] = 8        # add a new fruit
fruits["banana"] = 7        # update existing fruit
print("After adding/updating fruits:", fruits)

# 4. Removing elements
fruits.pop("orange")        # remove a fruit
print("After removing orange:", fruits)

# 5. Keys, values, items
print("Fruits available:", fruits.keys())
print("Quantities:", fruits.values())
print("Fruits and quantities:", fruits.items())
# 7. Iteration
print("Looping through fruits:")
for fruit, quantity in fruits.items():
    print(fruit, ":", quantity)
```
